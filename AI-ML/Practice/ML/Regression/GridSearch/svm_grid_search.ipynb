```python
import pandas as pd

dataset = pd.read_csv("insurance_pre.csv")

dataset = pd.get_dummies(dataset, drop_first=1)

#dataset.columns

independent_data = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent_data = dataset[['charges']]

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(independent_data, dependent_data)

# preprocessing
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR

grid_param = {'kernel': ['rbf', 'poly', 'sigmoid', 'linear'], 
              'C':[10,100,1000,2000,3000], 
              'gamma':['auto', 'scale']}

grid = GridSearchCV(SVR(), grid_param, refit = True, verbose = 3, n_jobs=-1)

grid.fit(X_train, Y_train)

print("Best Parammeter: ", grid.best_params_)
```

```python
re = grid.cv_results_

table = pd.DataFrame.from_dict(re)
table
```
