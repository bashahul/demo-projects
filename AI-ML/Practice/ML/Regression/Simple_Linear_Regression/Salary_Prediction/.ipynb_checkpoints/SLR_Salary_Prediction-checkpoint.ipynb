```python
# import lib to process dataset
import pandas as pd 
```

```python
# read data from csv
dataset = pd.read_csv("Salary_Data.csv") 
dataset
```

```python
# split data into input and output
independent_dataset = dataset[["YearsExperience"]]
dependent_dataset = dataset[["Salary"]]
independent_dataset
dependent_dataset
```

```python
# split the data into training and test set using sklearn
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(independent_dataset, dependent_dataset, test_size=0.30, random_state=0)
x_train
x_test
y_train
y_test
```

```python
# learning phase using sklearn's LinearRegression class
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(x_train, y_train)
weight = regressor.coef_
bias = regressor.intercept_

weight
bias
```

```python
# predict
y_predict = regressor.predict(x_test)

y_predict
```

```python
# evaluate model by R2 Score algorithm
from sklearn.metrics import r2_score
r_score = r2_score(y_test, y_predict)
r_score
```

```python
# save and load model for prediction
import pickle
file_name = "finalized_model_linear"

pickle.dump(regressor, open(file_name, 'wb'))

saved_model = pickle.load(open(file_name, 'rb'))

saved_model.predict(pd.DataFrame({'YearsExperience':[12]}))
```
