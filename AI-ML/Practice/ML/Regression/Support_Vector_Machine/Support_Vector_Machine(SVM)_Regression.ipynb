```python
# import library to process dataset and read csv
import pandas as pd

dataset = pd.read_csv("50_Startups.csv")
dataset.columns
```

```python
# Preprocess - Nominal Categorical data into numerical
dataset = pd.get_dummies(dataset, drop_first=1)
dataset.columns
```

```python
# split input and output
independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend',
       'State_Florida', 'State_New York']]
dependent = dataset[['Profit']]
```

```python
# split train and test set
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(independent, dependent)
```

```python
# standardize the data to reduce the difference data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

sc.fit_transform(x_train)
sc.transform(x_test)
```

```python
# train the model now
from sklearn.svm import SVR

regressor = SVR(kernel='rbf', C=1000)

regressor.fit(x_train, y_train)
```

```python
# predict
y_predict = regressor.predict(x_test)
```

```python
# evaluate model
from sklearn.metrics import r2_score

r_score = r2_score(y_test, y_predict)

r_score
```

```python
# save the model
import pickle as pk

file_name = "finalized_profit_prediction_svm_model.sav"

pk.dump(regressor, open(file_name, 'wb'))
```
